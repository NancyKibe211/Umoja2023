```python
# Import libraries
import requests
from tqdm.auto import tqdm
import shutil
# Function to download data
def zindi_data_downloader(url, token, file_name):
    # Get the competition data
    competition_data = requests.post(url = url, data= token, stream=True)
    
    # Progress bar monitor download
    pbar = tqdm(desc=file_name, total=int(competition_data.headers.get('content-length', 0)), unit='B', unit_scale=True, unit_divisor=512)
    # Create and Write the data to colab drive in chunks
    handle = open(file_name, "wb")
    for chunk in competition_data.iter_content(chunk_size=512): # Download the data in chunks
        if chunk: # filter out keep-alive new chunks
                handle.write(chunk)
        pbar.update(len(chunk))
    handle.close()
    pbar.close()
    
# Data url, token and file_name
data_urls = [
    "https://api.zindi.africa/v1/competitions/umojahack-africa-2023-beginner-challenge/files/SampleSubmission.csv",
             "https://api.zindi.africa/v1/competitions/umojahack-africa-2023-beginner-challenge/files/Test.csv",
             "https://api.zindi.africa/v1/competitions/umojahack-africa-2023-beginner-challenge/files/Train.csv",
            #  "https://api.zindi.africa/v1/competitions/wadhwani-ai-bollworm-counting-challenge/files/images_bboxes.csv",
#              "https://api.zindi.africa/v1/competitions/wadhwani-ai-bollworm-counting-challenge/files/SampleSubmission.csv"
]# url
token = {'auth_token': 'zus.v1.xqSYD11.EwqAXuLdkuBgr6TokGKBaDA7Vu9kU4'} # Use your own token
file_names = [x.split("/")[-1] for x in data_urls]
# Download data
[zindi_data_downloader(url = data_url, token = token, file_name = file_name) for data_url, file_name in zip(data_urls, file_names)]
# Unzip audio zip file
data_path = '.'
# shutil.unpack_archive(data_path + '/images.zip', 'images')
```
